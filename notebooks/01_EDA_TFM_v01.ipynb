```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os 

# Configuración visual para gráficos académicos
sns.set_theme(style="whitegrid", context="paper")

# 1. Cargar los datos
# Nota: Ajustad la ruta relativa según vuestra estructura de carpetas.
# Según la muestra, el delimitador es punto y coma (;)
ruta_archivo = '../data/raw/02-2026.csv'
df = pd.read_csv(ruta_archivo, sep=';')

# 2. Convertir la columna 'fecha' a datetime
# El formato en el subset es "YYYY-MM-DD HH:MM:SS"
df['fecha'] = pd.to_datetime(df['fecha'])

# 3. Filtrar datos desde 2022 en adelante (Mitigación de Concept Drift por normativa ZBE)
df_post2022 = df[df['fecha'] >= '2022-01-01'].copy()

# 4. Establecer la fecha como índice temporal
df_post2022.set_index('fecha', inplace=True)

# 5. Ordenar cronológicamente (Prevención de Data Leakage)
df_post2022.sort_index(inplace=True)

# 6. Primer vistazo a la estructura
print("Formato del dataset filtrado:", df_post2022.shape)
display(df_post2022.head())
print("\n--- Información de Tipos y Nulos ---")
df_post2022.info()
```

```python
# --- SANITY CHECK: Comprobación de los límites temporales ---

# Extraemos la fecha mínima y máxima directamente del índice temporal
fecha_minima = df_post2022.index.min()
fecha_maxima = df_post2022.index.max()
total_registros = len(df_post2022)

print(f"Fecha mínima en el dataset filtrado: {fecha_minima}")
print(f"Fecha máxima en el dataset filtrado: {fecha_maxima}")
print(f"Total de registros: {total_registros:,}") # El formato :, añade separadores de miles para mejor lectura
```
