```python
import pandas as pd

# 1. Cargar el dataset
# Nota: Pandas necesita el motor 'pyarrow' o 'fastparquet' para leer este formato.
# Si te da error, instala en tu terminal: pip install pyarrow
ruta_archivo = '../data/processed/dataset_completo.parquet'
df_m30 = pd.read_parquet(ruta_archivo)

# 2. Inspección de tipos de datos y memoria
print("--- 1. Información General del Dataset ---")
df_m30.info()
print("\n" + "="*50 + "\n")

# 3. Verificación visual del inicio y fin de la serie
print("--- 2. Primeros y últimos registros ---")
display(df_m30.head())
display(df_m30.tail())

# 4. Detección de valores nulos explícitos
print("\n--- 3. Conteo de Valores Nulos ---")
print(df_m30.isnull().sum())

# 5. Estadísticas descriptivas para detectar anomalías (ej. intensidades negativas)
print("\n--- 4. Estadísticas Descriptivas ---")
display(df_m30.describe())
```
